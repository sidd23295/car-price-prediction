```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv(r'cardataset/data.csv')
len(df)
```

```python
df.head(50)
```

```python
### removing spaces from the columns and replacing them with underscores
df.columns = df.columns.str.lower().str.replace(' ','_')
columns_strings = list(df.dtypes[df.dtypes=='object'].index)
columns_strings
```

```python
for col in columns_strings:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```

```python
df.head()
```

```python
## Our target is MSRP - we need to do target variable analysis next
sns.histplot(df.msrp, bins=40)
```

```python
sns.histplot(df.msrp[df.msrp<100000])
```

```python
## To remove the long tail we can make apply the log to the price
log_price = np.log1p(df.msrp)
sns.histplot(log_price)
```
