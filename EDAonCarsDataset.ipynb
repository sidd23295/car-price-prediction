```python
### splitting data into training, validation and testing data
totalrows = len(df)
n_val = int(0.2 * totalrows)
n_test = int(0.2 * totalrows)
n_train = totalrows - n_test - n_val

np.random.seed(2)
index = np.arange(totalrows)
np.random.shuffle(index)

df_shuffled = df.iloc[index]

df_train = df_shuffled.iloc[:n_train].copy()
df_val = df_shuffled.iloc[n_train:n_train + n_val].copy()
df_test = df_shuffled.iloc[n_train + n_val:].copy()

y_train = np.log1p(df_train.msrp.values)
y_test = np.log1p(df_test.msrp.values)
y_val = np.log1p(df_val.msrp.values)

del df_train['msrp']
del df_val['msrp']
del df_test['msrp']
```

```python
def train_linregression(X,y):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])
    
    XTX = X.T.dot(X)
    XTX_inv = np.linalg.inv(XTX)
    w = XTX_inv.dot(X.T).dot(y)
    return w[0], w[1:]
```

```python
columns = ['engine_hp', 'engine_cylinders', 'highway_mpg', 'city_mpg', 'popularity']
df_numerical = df_train[columns]
df_numerical.fillna(0, inplace=True)
```

```python
X_train = df_numerical.values
w_0, w = train_linregression(X_train, y_train)
```

```python
y_pred = w_0 + X_train.dot(w)
```

```python
sns.histplot(y_pred, label='prediction')
sns.histplot(y_train, label='target')
plt.legend()
```
